## HW1-5. category_id별 가장 Hot한 분야

```python
ids = CA.category_id.unique() 
print(sorted(ids))
print(len(sorted(ids)))
```

```python
CA_01=CA.loc[:, ['category_id', 'views', 'likes', 'dislikes', 'comment_count']]
CA_01.head()
```

```python
group1 = CA_01.groupby('category_id').views.sum()
print(group1)
# 조회수는 24:Entertainment 분야가 가장 많다.
```

```python
group2 = CA_01.groupby('category_id').likes.sum()
print(group2)
# 좋아요수는 10:Music 분야가 가장 많다.
```

```python
group3 = CA_01.groupby('category_id').dislikes.sum()
print(group3)
# 싫어요수는 24:Entertainment 분야가 가장 많다.
```

```python
group4 = CA_01.groupby('category_id').comment_count.sum()
print(group4)
# 코멘트수는 24:Entertainment 분야가 가장 많다.
```

```python
groups = pd.DataFrame([group1, group2, group3, group4])
groups
```

결론: category_id별로 분석한 결과 조회수, 코멘트수는 24:Entertainment 분야가 가장 많았다.  
또한 좋아요를 가장 많이 받은 분야는 10:Music이며, 싫어요를 가장 많이 받은 분야는 조회수가 가장 많았던 24:Entertainment였다.
